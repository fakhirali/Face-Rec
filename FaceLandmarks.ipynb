```python
def on_head(img,mask, face_landmarks_list):
    left_most = face_landmarks_list[0]['chin'][0][0]
    right_most = face_landmarks_list[0]['chin'][-1][0]
    how_high = face_landmarks_list[0]['left_eyebrow'][0][1] - face_landmarks_list[0]['nose_tip'][0][1]
    how_high = face_landmarks_list[0]['left_eyebrow'][0][1] - how_high
    #print(how_high,mask.shape[0],left_most,right_most)
    mask = cv2.resize(mask,(left_most-right_most,int((left_most-right_most)/mask.shape[0])*mask.shape[1]))
    print(mask.shape)
    print(img.shape)
    img[how_high:mask.shape[0] , left_most:right_most] = mask
    return img
```
